```python
df = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/AIM/dialect_dataset.csv')
df.head()
```

```python
id_list = list(df.id.astype(str))
len(id_list)
```

```python
q=0
s1 = pd.DataFrame()
for i in range(1,int(len(id_list)/1000) + 1):
    j = i *1000
    x = requests.post(url, json = id_list[q:j])
    s2 = pd.Series(x.json())
    s2 = s2.reset_index()
    s1 = s1.append(s2)
    q=j
    
```

```python
s1.columns = ["id", "txt"]
```

```python
df2 = df.copy()
df2['id'] = df['id'].astype(str)
df2
```
